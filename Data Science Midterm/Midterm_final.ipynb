# Red Wings CSV function

```python
def wings_csv(file, name):
    # drop unneeded columns from csv
    file = file.drop(columns = ['Unnamed: 6','Unnamed: 7','Unnamed: 13','Team.2','Team.3','Team.4','Oppo.2','Oppo.3','Oppo.4'])
    # rename columns
    file = file.rename(columns={'Unnamed: 0': 'Games Played', 'Unnamed: 1': 'Date', 'Unnamed: 2': 'Opponent', 
                                'Unnamed: 3': 'Red Wings: Goals For', 'Unnamed: 4': 'Red Wings: Goals Against',
                                'Unnamed: 5': 'Win or Loss', 'Team': 'Red Wings: Shots on Goal', 'Team.1': 'Red Wings: Penalty Minutes',
                               'Oppo': 'Opponent: Shots on Goal','Oppo.1': 'Opponent: Penalty Minutes'})
    # drop first row from csv
    file = file.drop(0)
    # change data type to float
    file['Red Wings: Goals For'] = file['Red Wings: Goals For'].astype(float)
    file['Red Wings: Goals Against'] = file['Red Wings: Goals Against'].astype(float)
    file['Red Wings: Shots on Goal'] = file['Red Wings: Shots on Goal'].astype(float)
    file['Red Wings: Penalty Minutes'] = file['Red Wings: Penalty Minutes'].astype(float)
    file['Opponent: Shots on Goal'] = file['Opponent: Shots on Goal'].astype(float)
    
    # save to cleaned csv file
    file = file.to_csv(f'{name}.csv', index = False)
```

# Golden Knights CSV function

```python
def gold_csv(file, name):
    # drop unneeded columns from csv
    file = file.drop(columns = ['Unnamed: 7','Unnamed: 13','Unnamed: 6','Team.2','Team.3','Team.4','Oppo.2','Oppo.3','Oppo.4'])
    # rename columns
    file = file.rename(columns={'Unnamed: 0': 'Games Played','Unnamed: 1': 'Date', 'Unnamed: 2': 'Opponent', 
                                'Unnamed: 3': 'Golden Knights: Goals For', 'Unnamed: 4': 'Golden Knights: Goals Against',
                                'Unnamed: 5': 'Win or Loss', 'Team': 'Golden Knights: Shots on Goal', 
                                'Team.1':'Golden Knights: Penalty Minutes','Oppo': 'Opponent: Shots on Goal','Oppo.1': 'Opponent: Penalty Minutes'})
    # drop first row from csv
    file = file.drop(0)
    file['Golden Knights: Goals For'] = file['Golden Knights: Goals For'].astype(float)
    file['Golden Knights: Goals Against'] = file['Golden Knights: Goals Against'].astype(float)
    file['Golden Knights: Shots on Goal'] = file['Golden Knights: Shots on Goal'].astype(float)
    file['Golden Knights: Penalty Minutes'] = file['Golden Knights: Penalty Minutes'].astype(float)
    file['Opponent: Shots on Goal'] = file['Opponent: Shots on Goal'].astype(float)
    file['Opponent: Penalty Minutes'] = file['Opponent: Penalty Minutes'].astype(float)
    
    # save to cleaned csv file
    file = file.to_csv(f'{name}.csv', index = False)
```

## Web Scraping Function

```python
def scrape(file, class_name):
    header = []
    rows = []
    # find table object
    table = file.find('table', class_= class_name)
    header = []
    rows = []
    # iterate through each row and extract either header or row values:
    for i, row in enumerate(table.find_all('tr')):
        if i == 0:
            header = [el.text.strip() for el in row.find_all('th')]
        else:
            rows.append([el.text.strip() for el in row.find_all('td')])
    return header, rows
```

My favorite sport is hockey and my favorite team is the Detroit Red Wings in the National Hockey league. They are currently in a rough patch/in a period of rebuilding. I was curious to see if we actually have been improving at all the last five seasons and also how do our last five seasons compare to last 5 seasons of the 2022-23 season's Stanley Cup winner, the Las Vegas Golden Knights. I also wanted to compare the last 5 top players (determined by who ended the season with the most points) on our team and theirs along with the last 5 starter and backup goalies (determined by the two goalies who played the most games). The data per season that I wanted to pay close attention to was goals scored, goals let in/scored against, shots on goal (meaning how many times is the team shooting the puck), and penalty minutes. There is so much that goes into what makes a team good or bad team but I had to have a limit to what I was looking at or it would have been an overwhelming amount of data. To win a game you need to be making goals, how many you let in can lose you the game, if you aren't taking shots it's much harder to score and how many shots can be a sign of good or bad offense, and if you are getting loads of penalties you are giving the opposing team more opportunities to get a power play which means we play with one less man than them and makes the game easier for the opposing team. For the players I wanted to look at how many goals they were scoring, how many assists (being the person that passes the puck to the person who scores), how many points they got, and their plus minus rating (you get a plus by being on the ice when you team scores a goal and you get a minus when you are on the ice when your team lets in a goal and those are averaged out per season and ideally you will be in the plus and not minus). For the goalies I looked at goals against average (the number of goals a goaltender allows per 60 minutes of playing time. Calculated by the number of goals against multiplied by 60 (minutes) and then dividing by number of minutes played), number of wins, and number of losses. 

Unfortunately the last 5 seasons does include the "covid seasons" which were not full seasons so I'm sure that did have an affect on the data but luckily both teams would have had their data affected in the same way and when calculating the mean of the team data it did take into account the lower number of games and should still be a comparable mean to the other full seasons. I found csv files online for the full team season data but the files were not that great so I had to clean them up a lot to be readable but also only include relevant data. I created a function for the Red Wings files and the Golden Knights files since I was cleaning them in the same way. The function made it much easier and significantly less code; you just had to input the csv file you wanted to clean and a name for the cleaned csv file when saving it. I used pandas, csv, and matplotlib to work with the files and to graph them.
